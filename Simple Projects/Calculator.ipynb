# Simple Calculator Project - Practical Notes

## Project Goal
This notebook implements a basic calculator that performs addition, subtraction, multiplication, and division based on user input.

## Key Concepts Used

### 1. Variables and Assignment Operator (`=`)
- The single `=` is the **assignment operator**. 
- It is used to store a value into a variable.
- Example: `operator = input(...)` assigns the user's input to the variable.

### 2. Comparison Operator (`==`)
- The double `==` is the **equality comparison operator**.
- It checks whether two values are equal.
- Important distinction: `=` assigns, `==` compares.

### 3. Conditional Logic: if / elif / else
- `if` statements allow the program to execute code only when a condition is True.
- `elif` (else if) checks additional conditions if the previous ones were False.
- Used here to branch execution based on the chosen arithmetic operator.
- This is the standard way to handle multiple possible paths in Python.

### 4. Arithmetic Operators
- `+` → Addition
- `-` → Subtraction
- `*` → Multiplication
- `/` → Division (always returns float)

### 5. Input Handling
- `input()` always returns a string.
- We convert numbers using `float()` to support decimal results in division.

```python
#this is a message or guideline to the user before the program starts
print(f"Use", "+",  "for addition")
print(f"Use", "-",  "for subtraction")
print(f"Use", "*",  "for matplication")
print(f"Use", "/",  "for division")

#This is the first variable and we assign the value ( input function) so that we can store the oparator)
oparator = input("Choose an oparator: ")

#These 2 variables store the values of number in float type so that we can do math.
#It stores the numbers chosen by the user (from the input function).
#input function always returns a string and we convert it to float ( a decimal number ) so that we can work with devisions also
num1 = float(input("Enter a number: "))
num2 = float(input("Enter a number: "))


#The if/else logic checks the oparator ( from user's input) and the 2 numbers( also from user's input) and does the math
# the double == is the equality comparision operator used to check if the values of two objects are equal 
#NOTE: a single = is the assigment operator which is used to store a value insside a variable 
if oparator == "+":
    print(num1 + num2)

elif oparator == "-":
    print(num1 - num2)

elif oparator == "*":
    print(num1 * num2)

elif oparator == "/":
    print(num1 / num2)
```
