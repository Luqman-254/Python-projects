## What Is a Text Adventure?

A text adventure is a game where the player interacts with the world through text.

Instead of controlling a character with graphics, the player reads descriptions and chooses actions.

For example:

1. The game describes a situation.
2. The player chooses an option.
3. The story changes based on that choice.
4. The process repeats until an ending is reached.

Many early computer games were built using this approach.

# Conditional Statements in Python

## Introduction

Programs become interesting when they can make decisions.

Without decision-making, a program simply executes instructions from top to bottom.

Conditional statements allow a program to choose different paths depending on information provided by the user or calculated by the program.

---

# The `if` Statement

The `if` statement is used when code should execute only when a condition is true.

```python
age = 18

if age >= 18:
    print("You can vote.")
```

Python evaluates the condition.

If the condition is true, the indented block runs.

If the condition is false, Python skips the block.

---

# The `else` Statement

The `else` statement provides an alternative path.

```python
age = 15

if age >= 18:
    print("You can vote.")
else:
    print("You cannot vote yet.")
```

Only one of the two blocks will execute.

---

# The `elif` Statement

Sometimes there are multiple possibilities.

The `elif` statement allows additional conditions to be checked.

```python
score = 85

if score >= 90:
    print("Grade A")
elif score >= 80:
    print("Grade B")
else:
    print("Grade C")
```

Python checks conditions from top to bottom.

The first condition that evaluates to `True` is executed.

---

Every conditional statement asks a question.

---

# Nested `if` Statements

A nested `if` statement is an `if` statement placed inside another `if` statement.

The second decision only becomes available after the first decision has been made.

Example:

```python
has_ticket = True
has_id = True

if has_ticket:
    if has_id:
        print("Enter airport")
```

The second condition is checked only if the first condition is true.

# Why Nested Conditionals Matter

Nested conditionals are useful whenever a decision depends on a previous decision.

Common examples include:

* Adventure games
* Menu systems
* Login systems
* Access control
* User validation

Many real-world programs rely on nested decision-making.
